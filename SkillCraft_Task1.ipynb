**Importing the libraries required**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Loading the dataset**

```python
df = pd.read_csv('DataPopulation.csv', skiprows=4)
```

**Selecting a subset of the dataframe for specific year and specific country**

```python
selected_countries = ['United States', 'China', 'India', 'Brazil', 'Germany']
df_2020 = df[df['Country Name'].isin(selected_countries)][['Country Name', '2020']]
```

**Renaming columns for easier access**

```python
df_2020.columns = ['Country', 'Population_2020']
```

**Creating the bar chart for population in 2020 ( Population Distribution by Year)**

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Country', y='Population_2020',color='Green', data=df_2020)
plt.title('Population of Selected Countries in 2020')
plt.xlabel('Country')
plt.ylabel('Population (2020)')
plt.xticks(rotation=45)
plt.show()
```

**Histogram for Distribution of Population in 2020**

```python
plt.figure(figsize=(10, 6))
plt.hist(df['2020'].dropna(), bins=30,color='red', edgecolor='black')
plt.title('Distribution of Population Across Countries in 2020')
plt.xlabel('Population (2020)')
plt.ylabel('Frequency')
plt.show()
```
